```python
import pickle
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict
from sklearn.preprocessing import StandardScaler

def compare(subA, subB, scale=False):
    pols_A = pickle.load(open("data/{}/polarities.pkl".format(subA), "rb"))
    pols_B = pickle.load(open("data/{}/polarities.pkl".format(subB), "rb"))

    means_A = defaultdict(list)
    means_B = defaultdict(list)
    for run in pols_A:
        for word, value in run.items():
            means_A[word].append(value)
            
    for run in pols_B:
        for word, value in run.items():
            means_B[word].append(value)

    df_A = pd.DataFrame([(word, np.nanmean(values), np.nanstd(values)) for word, values in means_A.items()], columns =["word", "mean", "std"])
    df_B = pd.DataFrame([(word, np.nanmean(values), np.nanstd(values)) for word, values in means_B.items()], columns =["word", "mean", "std"])

    df_A = df_A.sort_values(by=["mean"])
    df_B = df_B.sort_values(by=["mean"])

    if scale:
        scaler = StandardScaler(with_std=False)
        df_A["mean"] = scaler.fit_transform(df_A["mean"].values.reshape(-1, 1))
        df_B["mean"] = scaler.fit_transform(df_B["mean"].values.reshape(-1, 1))

    joined = df_A.merge(df_B, how="inner", on="word", suffixes=("_" + subA, "_" + subB))

    joined["difference"] = np.abs(joined["mean_" + subA].values - joined["mean_" + subB].values)

    joined = joined.sort_values(by=["difference"])
    return df_A, df_B, joined

```

```python
A, B, joined = compare("texas", "nyc")
```
